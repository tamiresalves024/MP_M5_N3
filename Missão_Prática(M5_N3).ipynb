```python
import pandas as pd

csv_file_path = '/data.csv'

data = pd.read_csv(csv_file_path, sep=';', engine='python', encoding='utf-8')

data_copia = data.copy()

data_copia['Calories'].fillna(0, inplace=True)
print(data_copia)
```

```python
data_copia['Date'].fillna('1900/01/01', inplace=True)
print(data_copia)
```

```python
data_copia['Date'].replace('1900/01/01', pd.NaT, inplace=True)
data_copia['Date'] = pd.to_datetime(data_copia['Date'], errors='coerce')
print(data_copia)
```

```python
data_copia['Date'] = data_copia['Date'].astype(str).str.replace('20201226', '2020-12-26')
data_copia['Date'] = pd.to_datetime(data_copia['Date'], errors='coerce')
print(data_copia)
```

```python
data_copia.dropna(inplace=True)
print(data_copia)
```
